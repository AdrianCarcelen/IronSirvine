# How to webscrap with python

```python
import requests # Import library
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/2021_in_film' 
html = requests.get(url).content # This is how we get the information from the webside
html[0:600]
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html")
soup
```

```python
tags = ['h2']
text=[]
for element in soup.find_all(tags):
    text.append(element.text.replace('[edit]',''))
text
```

```python
info_box = soup.find('table',class_="infobox")

print(info_box.text)
```

```python
movies_table = soup.find_all('table', class_='wikitable')
# print(movies_table[2].text)

tables = soup.find_all('table', attrs={'class':'wikitable'})

fest_table = None
for table in tables:
    cap = table.find('caption')
    if cap is None:
        continue
    if 'film festivals' in cap.text:
        # print(table.text)
        fest_table = table

body_fest_table = fest_table.find('tbody')

line1 = body_fest_table.find_all('tr')[1]

d,e,h,l = line1.find_all('td')
body_fest_table.find_all('tr')
rows = body_fest_table.find_all('tr')
```

```python
line0 = body_fest_table.find_all('tr')[0]
d,e,h,l = line0.find_all('th')
print(d.text.strip(),e.text.strip(), \
      h.text.strip(),l.text.strip(), sep=';')
print('-;-;-;-')
for line in body_fest_table.find_all('tr')[1:]:
    d,e,h,l = line.find_all('td')
    print(d.text.strip(),e.text.strip(), \
          h.text.strip(),l.text.strip(), sep=';')
```

```python
rows = body_fest_table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows] # best way more fast an ez

colnames = rows[0]
data = rows[1:]
 
miau = pd.DataFrame(data, columns=colnames)
miau
```
