# MatPlotLib

```python
import matplotlib.pyplot as plt

def f(x):
    return 0.5*x*x-3
def g(x):
    return 40-x*x

plt.rcParams['figure.figsize'] = [6,4]
xs = np.linspace(-10,10,50)
ys = f(xs)
zs = g(xs)
plt.plot(xs,ys, '+', color=(1,0.5,0), label='I am f')
plt.plot(xs,zs, '+', color=(0,0.5,1), label='I am g')
plt.legend()
# plt.savefig('myplot.svg')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
males = df[df['Sex']=='male']
females = df[df['Sex']=='female']
```

```python
a = np.array(males['Age'])
y = np.ones_like(a)
plt.scatter(a,y, label='males')
a = np.array(females['Age'])
y = np.zeros_like(a)
plt.scatter(a,y, label='females')
plt.legend()
```

```python
xs = np.linspace(-2,2, 100)
ys = np.linspace(-2,2, 100)
def f(x,y):
    return 2**(-x*x-y*y)
# make a grid of xs and ys
X,Y = np.meshgrid(xs,ys)
Z = f(X,Y)
plt.imshow(Z, cmap='jet')
```

```python
# plot a 3d surface
from mpl_toolkits.mplot3d import axes3d
ax = plt.figure().add_subplot(projection='3d')
ax.plot_surface(X,Y,Z, cmap='jet', alpha=0.3)
ax.contour(X, Y, Z, zdir='z', offset=-0.5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=2, cmap='coolwarm')
ax.set(xlim=(-2, 2), ylim=(-2, 2), zlim=(-0, 1),
       xlabel='X', ylabel='Y', zlabel='Z')
plt.show()
```
