```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def throw(n):
    s = 0
    for i in range(n):
        if random.random() > 0.5:
            s += 1
        else:
            s += 0
    return s
```

```python
l = []
n_repeat = 100000
n_dices = 100
for j in range(n_repeat):
    l.append(throw(n_dices))
```

```python

plt.hist(l, bins=n_dices)
plt.show()
```

```python
# (np.random.random(n_dices)<0.5).sum()
l = np.sum(np.random.random((n_dices, n_repeat))<0.5, axis=0)

plt.hist(l, bins=n_dices)
plt.show()
```
