# GAME FUNCTIONS
### Here we put all the functions needed to run the Game Smoothly

```python
#Type function OK
def input_type(list_to_show)->str:
    print("""Type "ONE" of theses words : """, *list_to_show,'\n')
    typed_text = input(list_to_show)
    return typed_text.lower()

#Check input
def check_input(input_typed: str, wanted_type: str)->bool:
    check_presence = input_typed in wanted_type
    return check_presence

#Open door check function
def check_access(door_to_open: str, inventory: set):
    """This function checks if the Player has access to the next room/
    by checking the presence of the 'Key' corresponding 'door'/
    Ultimately it will update the player_position.

    Args:
        door_to_open (str): Name of the corresponding door
        inventory (set): Parsed to verify if the key is present
    """
    global player_position
    if dict_interact[door_to_open] in inventory:
            player_position = dict_doors[player_position][door_to_open]
            print(f"You open the {door_to_open} and enter the {player_position}.")
    else:
        print("Hey! you don't have the key to access that room.\n")
        return

def find_object(object_to_check: str, inventory: set, examine_dict: dict):
    """This function is used to look for the hidden objects in the room items\
    and to reroute to check_access in case player want to examine the doors.

    Args:
        object_to_check (str): object selected used to check if the player selected the item\n
        that has an hidden object
        inventory (set): parsed to check if objet already present in inventory
        examine_dict (dict): used to verify / add new item in player's inventory
    """
    for item in examine_dict:
        if examine_dict[object_to_check] == "It's a door.":
            print(examine_dict[object_to_check])
            check_access(object_to_check, inventory)
            return
        else:
            if object_to_check != item:
                object_to_add = examine_dict[object_to_check]
                inventory.add(object_to_add)
                print(object_to_add, " added to your inventory.")
            else:
                print("Oh, I think there's nothing here.")

print(dict_doors[player_position]['door a'])
print(set_inventory)
```

# GAME MECANICS
### Here we put the game logic by calling the Game Functions

```python
# Start in game room
print(*start_text)
print("Hey fellow adventurer type what you want to do : ", tuple_actions)

while player_position != 'You Win!':
    """
    The player_position define in witch room we are.

    The the switch_action is use to give two choices to the player.\
    Either exploring, or investigate the objects of the room.

    If exploring, we'll check the inventory
    """
    print(dict_welcome_text[player_position])#Could add function to customise text depending of the room
    switch_action = ''

    while not check_input(switch_action, tuple_actions):#checking if typed text is in tuple_action
        switch_action = input_type(tuple_actions)#switch for choose between examine and interact

    if switch_action == 'examine':#examine/object case
        examine_keys = dict_examine[player_position]
        print('So.. feeling lucky?In the ', player_position, ' you see ', examine_keys.keys())
        object_chosen = input_type(examine_keys.keys())
        if check_input(object_chosen, examine_keys.keys()):
            #object_to_inventory = select_door(object_chosen, set_inventory, examine_keys.keys())
            find_object(object_chosen, set_inventory, examine_keys)

    elif switch_action == 'interact':#interact/doors case
        interact_keys = dict_doors[player_position]
        print("Your will to get out is crushing, you need to open the damn doors.")
        print("Witch one do you want to open :", *interact_keys.keys())
        door_chosen = input_type(interact_keys.keys())
        if check_input(door_chosen, interact_keys.keys()):
            check_access(door_chosen, set_inventory)

    else:
        print("Houston we have a problem!")

print(end_text)
print("""You win\nT H E   E N D""")

```
