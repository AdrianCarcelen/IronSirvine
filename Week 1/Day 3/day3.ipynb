# How to enter next step to iterations

```python
lst = []

for i in range(10,31,2):
    lst.append(i)

print(lst)
```

```python
lst = [i for i in range(10,31,2)] # we can make for sentences inside the list

print(lst)
```

```python
lst = [i for i in range(21) if i%5 != 0] # if u wana make a condition u need to put the if before the sentence

print(lst) 
```

```python
lst = [i if i%5 != 0 else 0 for i in range(21)] # if u also want to make a else condition need to do the if at the start of the sentence and then write the else sentence

print(lst) 
```

```python
lst = [j for i in range(5) for j in range(3)]

print(lst)
```

```python
lst = [[j for i in range(5)] for j in range(3)]

print(lst)
```

```python
import matplotlib.pyplot as plt


lst = [(i,j) for i in range(2) for j in range(2)]
x  = []
y  = []
for number in lst:
    x.append(number[0])
    y.append(number[1])

print(x,y,lst)

plt.plot(lst)
plt.show()

```

```python
lst = [(i,j,z) for i in range(2) for j in range(2) for z in range(2)]

print(lst)

plt.plot(lst)
plt.show()

```
